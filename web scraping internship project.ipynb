###### Q2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#and make data frame

```python
from bs4 import BeautifulSoup
import requests
```

```python
page= requests.get('https://www.imdb.com/chart/top')
page
```

```python
soup= BeautifulSoup(page.content)
soup
```

```python
!pip install html5lib
```

```python
soup= BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
Scraped_movies= []
for i in soup.find_all('td',class_='titleColumn'):
    Scraped_movies.append(i.text.replace('\n',""))
Scraped_movies
```

```python
Scraped_movies= Scraped_movies[0:100]
Scraped_movies
```

```python
# scrap ratingof movies
```

```python
Scraped_rating= []
for j in soup.find_all('td',class_='ratingColumn imdbRating'):
    Scraped_rating.append(j.text.replace('\n',""))
Scraped_rating    
```

```python
Scraped_rating= Scraped_rating[0:100]
Scraped_rating
```

```python
# scrap the year of release
```

```python
Release_year=[]
for k in soup.find_all('span',class_='secondaryInfo'):
    Release_year.append(k.text.replace('\n',""))
Release_year    
```

```python
Release_year= Release_year[0:100]
Release_year
```

```python
print(len(Scraped_movies),len(Scraped_rating),len(Release_year))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'movie_name':Scraped_movies,'Rating':Scraped_rating,'Year of Release':Release_year})
df
```

```python
df.to_csv('IMDB top 100 movies.csv',index=False)
```

###### Q3 Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of
#release) and make data frame.


```python
pages= requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
pages
```

```python
Soup= BeautifulSoup(pages.content)
Soup
```

```python
Scraped_indian_movies=[]
for b in Soup.find_all('td',class_='titleColumn'):
    Scraped_indian_movies.append(b.text.replace('\n',""))
Scraped_indian_movies    
```

```python
Scraped_indian_movies= Scraped_indian_movies[0:100]
Scraped_indian_movies
```

```python
Scraped_indian_movies_rating=[]
for c in Soup.find_all('td',class_='ratingColumn imdbRating'):
    Scraped_indian_movies_rating.append(c.text.replace('\n',""))
Scraped_indian_movies_rating    
```

```python
Scraped_indian_movies_rating= Scraped_indian_movies_rating[0:100]
Scraped_indian_movies_rating
```

```python
movie_release_year=[]
for d in Soup.find_all('span',class_='secondaryInfo'):
    movie_release_year.append(d.text.replace('\n',""))
movie_release_year
```

```python
movie_release_year= movie_release_year[0:100]
movie_release_year
```

```python
import pandas as pd
```

```python
data= pd.DataFrame({'movies_name':Scraped_indian_movies,'Rating':Scraped_indian_movies_rating,'Release_Year':movie_release_year})
data
```

```python
data.to_csv('IMDB top 100 indian movies.csv',index=False)
```
