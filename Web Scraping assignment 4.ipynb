1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
import pandas as pd
import selenium
from selenium import webdriver
import sys
import time
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
```

```python
# Connect to the webdriver
driver= webdriver.Chrome(r"C:/Users/LENOVO/Downloads/chromedriver_win32/chromedriver")
```

```python
driver.get(' https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
title_tag= driver.find_elements_by_tag_name("tr")
title_tag
```

```python
video_title=[]
```

```python
#extract the text
for i in title_tag:
    video_title.append(i.text)
video_title
```

```python
df1 = pd.DataFrame({'title':video_title})
```

```python
df1
```

```python
df1.to_csv('most watched video',index=False)
```

2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1
st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code

```python
# Connect to the webdriver
driver2= webdriver.Chrome(r"C:/Users/LENOVO/Downloads/chromedriver_win32/chromedriver")
```

```python
driver2.get('https://www.bcci.tv/international/fixtures')
```

```python
match_tag = driver2.find_elements_by_xpath("//span[@class='ng-binding']")
match_tag
```

```python
match=[]
for i in match_tag:
    match.append(i.text)
match    
```

```python
place_tag = driver2.find_elements_by_xpath("//span[@class='ng-binding ng-scope']")
place_tag
```

```python
place=[]
for a in place_tag:
    place.append(a.text)
place  
```

```python
date_tag = driver2.find_elements_by_xpath("//div[@class='match-card-left match-schedule']")
date_tag
```

```python
date=[]
for b in date_tag:
    date.append(b.text)
date
```

```python
time_tag = driver2.find_elements_by_xpath("//h5[@class='text-right ng-binding']")
time_tag
```

```python
time=[]
for c in time_tag:
    time.append(c.text)
time
```

```python
# creating a DAtaFrame
df2 = pd.DataFrame({'Match':match,'Place':place,'Date':date,'Time':time})
df2
```
